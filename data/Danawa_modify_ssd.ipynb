```python
import pandas as pd
import numpy as np
import time
import re
```

```python
ssd = pd.read_csv('ssd_final.csv')
```

```python
col = 'ram_mounted'
ssd[col].unique()
```

```python
for i in range(len(ssd[col])):
    if pd.isnull(ssd[col][i]): ssd.loc[i, col] = 0
    else: ssd.loc[i, col] = 1
```

```python
for col in ssd.columns:
    for i in range(len(ssd[col])):
        if pd.isnull(ssd[col][i]): continue
        if '최대 ' in str(ssd.loc[i, col]):
            print(col)
```

```python
cols = ['sequential_read', 'sequential_write']
for col in cols:
    print(ssd[col].unique())
```

```python
for col in cols:
    for i in range(len(ssd[col])):
        if pd.isnull(ssd[col][i]): continue
        ssd.loc[i, col] = re.sub('(최대 |\,|\.)', '', ssd.loc[i, col])
    print(ssd[col].unique())
```

```python
col = 'ram_type'
ssd[col].unique()
```

```python
today = time.strftime('%y%m%d')
pd.DataFrame(ssd).to_csv(f'part_ssd{today}.csv', index=None)
```
