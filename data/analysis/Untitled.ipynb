```python
parts = np.array([])
print(parts)
parts_col = []
parts_data = []
for idx, i in enumerate(arrr):
    part_data = []  # 반복문 내에서 초기화
    response = requests.get(f"{base_url}/cpu/{i[0]}")
    response = response.json()
    if idx == 0:
        parts_col.extend(response.keys())
    part_data.extend(response.values())  # .values()를 사용하여 NumPy 배열의 값을 얻음
    
    response = requests.get(f"{base_url}/gpu/{i[1]}")
    response = response.json()
    if idx == 0:
        parts_col.extend(response.keys())
    part_data.extend(response.values())  # .values()를 사용하여 NumPy 배열의 값을 얻음
    
    response = requests.get(f"{base_url}/ram/{i[2]}")
    response = response.json()
    if idx == 0:
        parts_col.extend(response.keys())
    part_data.extend(response.values())  # .values()를 사용하여 NumPy 배열의 값을 얻음
    
    response = requests.get(f"{base_url}/ssd/{i[3]}")
    response = response.json()
    if idx == 0:
        parts_col.extend(response.keys())
    part_data.extend(response.values())  # .values()를 사용하여 NumPy 배열의 값을 얻음
    
    parts_data.append(part_data)

print(parts_data)    
print(parts_col)
```

```python
for i, item in enumerate(parts_data):
    if item[16] == '' or item[16] == None:
        parts_data[i][16] = None
    else:
        parts_data[i][16] = int(item[16][2:4])
new_parts_data = [[0 if value is None else value for value in row] for row in parts_data]
new_parts_data = np.array(new_parts_data)

for i, v in enumerate(new_parts_data):
    for idx, item in enumerate(v):
        print(type(item))
```
