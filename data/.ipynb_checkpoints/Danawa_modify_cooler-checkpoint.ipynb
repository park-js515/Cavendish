```python
import pandas as pd
import numpy as np
import time
import re
```

```python
cooler = pd.read_csv('cooler230919.csv')
```

```python
col = 'fan_size'
cooler[col].unique()
```

```python
for i in range(len(cooler)):
    if pd.isnull(cooler[col][i]): continue
    cooler.loc[i, col] = cooler.loc[i, col].replace(' ', '')
    if 'x' in cooler[col][i]:
        cooler.loc[i, col] = cooler.loc[i, col][:cooler.loc[i, col].index('x')]
    elif ',' in cooler[col][i]:
        cooler.loc[i, col] = cooler.loc[i, col][:cooler.loc[i, col].index(',')]
    elif '.' in cooler[col][i]:
        cooler.loc[i, col] = cooler.loc[i, col][:cooler.loc[i, col].index('.')]
```

```python
cooler[col].unique()
```

```python
col = 'fan_count'
cooler[col].unique()
```

```python
for i in range(len(cooler)):
    if pd.isnull(cooler[col][i]): continue
    if '팬미포함' == cooler[col][i]:
        cooler.loc[i, col] = '0'
```

```python
cooler[col].unique()
```

```python
col = 'airflow'
cooler[col].unique()
```

```python
for i in range(len(cooler)):
    if pd.isnull(cooler[col][i]): continue
    cooler.loc[i, col] = cooler.loc[i, col].replace(' ', '')
    if '최대' in cooler[col][i]:
        cooler.loc[i, col] = cooler.loc[i, col].replace('최대', '')
    elif cooler[col][i].endswith('CFM'):
        cooler.loc[i, col] = cooler.loc[i, col].replace('CFM', '')
    elif cooler[col][i].endswith('FM'):
        cooler.loc[i, col] = cooler.loc[i, col].replace('FM', '')
    elif '~' in cooler[col][i]:
        cooler.loc[i, col] = cooler.loc[i, col][cooler.loc[i, col].index('~')+1:]
```

```python
cooler[col].unique()
```

```python
col = 'tdp'
cooler[col].unique()
```

```python
for i in range(len(cooler)):
    if pd.isnull(cooler[col][i]): continue
    if cooler[col][i].startswith('TDP '):
        cooler.loc[i, col] = cooler.loc[i, col].replace('TDP ', '')
```

```python
cooler[col].unique()
```

```python
cols = ['radiator_length', 'radiator_thickness', 'hose_length']
for col in cols:
    print(cooler[col].unique())
    for i in range(len(cooler)):
        if pd.isnull(cooler[col][i]): continue
        cooler.loc[i, col] = cooler.loc[i, col].replace('mm', '')
```

```python
for col in cols:
    print(cooler[col].unique())
```

```python
col = 'as_years'
cooler[col].unique()
```

```python
for i in range(len(cooler)):
    if pd.isnull(cooler[col][i]): continue
    if cooler.loc[i, col] == '6~9':
        cooler.loc[i, col] = '6'
```

```python
cooler[col].unique()
```

```python
col = 'noise'
cooler[col].unique()
```

```python
for i in range(len(cooler)):
    if pd.isnull(cooler[col][i]): continue
    if cooler.loc[i, col] == '24.6, 22.6':
        cooler.loc[i, col] = '24.6'
```

```python
cooler[col].unique()
```

```python
today = time.strftime('%y%m%d')
pd.DataFrame(cooler).to_csv(f'cooler{today}.csv', index=None)
```
